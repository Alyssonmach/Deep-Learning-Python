### Separando os Atributos Previsores e Meta

```python
# separando os atributos previsores dos atributos meta
previsores = dataframe.iloc[:, 1:13].values
meta = dataframe.iloc[:, 0].values
```

```python
# visualizando o dataframe
dataframe
```

### Transformando os Atributos Categóricos em Numéricos

```python
# criando o objeto 'labelencoder_previsores'
labelencoder_previsores = LabelEncoder()
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 0] = labelencoder_previsores.fit_transform(previsores[:, 0])
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 1] = labelencoder_previsores.fit_transform(previsores[:, 1])
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 3] = labelencoder_previsores.fit_transform(previsores[:, 3])
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 5] = labelencoder_previsores.fit_transform(previsores[:, 5])
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 8] = labelencoder_previsores.fit_transform(previsores[:, 8])
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 9] = labelencoder_previsores.fit_transform(previsores[:, 9])
```

```python
# realizando a transformação dos atributos categóricos em numéricos
previsores[:, 10] = labelencoder_previsores.fit_transform(previsores[:, 10])
```

```python
# visualizando os atributos previsores
print(previsores)
```

### Transformando Valores Numéricos em Variáveis do Tipo Dummy

```python
# criando o objeto 'onehotencoder'
onehotencorder = ColumnTransformer(transformers=[("OneHot", OneHotEncoder(), 
                                                  [0,1,3,5,8,9,10])], remainder='passthrough')
```

```python
# transformando os atributos previsores em variávies do tipo dummy
previsores = onehotencorder.fit_transform(previsores).toarray()
```

```python
# visualizando a dimensão dos atributos previsores
previsores.shape
```

```python
# visualizando os atributos previsores
print(previsores)
```

```python
# visualizando os atributos meta
print(meta)
```

## Montando a Rede Neural

### Importação das Bibliotecas

```python
# importando da bibliotecas keras do python
from keras.models import Sequential
from keras.layers import Dense
```

### Estrutura da Rede Neural

```python
# criando o objeto 'regressor'
regressor = Sequential()
```

```python
# adicionando a primeira camada densa na rede neural
regressor.add(Dense(units = 158, activation = 'relu', 
                   input_dim = 316))
```

```python
# adicionando a segunda camada densa na rede neural
regressor.add(Dense(units = 158, activation = 'relu'))
```

```python
# adicionando a camada de saída na rede neural
regressor.add(Dense(units = 1, activation = 'linear'))
# como é um problema de regressão, a função de ativação não pode
# alterar o valor da saída
```

```python
# ajustando os parâmetros da rede
regressor.compile(loss = 'mean_absolute_error', optimizer = 'adam',
                 metrics = ['mean_absolute_error'])
```

### Treinando a Rede Neural

```python
# treinando a rede neural com o soatriutos previsores e meta
regressor.fit(previsores, meta, batch_size = 300, epochs = 500)
```

### Prevendo com a Rede Neural Criada

```python
# obtendo a previsão com os atributos previsores
previsoes = regressor.predict(previsores)
```

```python
# visualizando as previsoes
print(previsoes)
```

```python
# visualizando os atributos meta
print(meta)
```

Obsrve que os dados estão com um erro muito grande. Entretanto, a média dos dados tende a ser muito próxima.

```python
# obtendo a média dos atributos meta
meta.mean()
```

```python
# obtendo a média das previsões do algoritmo em relação aos atributos previsores
previsoes.mean()
```
