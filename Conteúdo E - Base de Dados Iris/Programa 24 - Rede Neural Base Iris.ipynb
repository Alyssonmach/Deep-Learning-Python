# Base de Dados Iris
#### Classificando uma iris em setosa, virginica ou versicolor.

### Importação das Bibliotecas

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

### Importando a Base de Dados

```
dataframe = pd.read_csv('/content/drive/My Drive/iris.csv', encoding = 'utf-8', sep = ',')
previsores = dataframe.iloc[:, 0:4].values
classe = dataframe.iloc[:, 4].values
```

### Pré-processamento com a Base de Dados

```
labelencoder = LabelEncoder()
classe = labelencoder.fit_transform(classe)
classe_dummy = np_utils.to_categorical(classe)
```

|**Tipo**|**Variável Dummy**|
|:------:|:----------------:|
|Setosa|100|
|Virginica|010|
|Versicolor|001|

```
x_train, x_test, y_train, y_test = train_test_split(previsores, classe_dummy, 
                                                     test_size = 0.25)
```

```
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Configuração da Rede Neural Artificial

```
classificador = Sequential()

classificador.add(Dense(units = 4, activation = 'relu', input_dim = 4))
classificador.add(Dense(units = 4, activation = 'relu'))
classificador.add(Dense(units = 3, activation = 'softmax', input_dim = 4))
```

```
classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy', 
                      metrics = ['categorical_accuracy'])
```

### Treinando a Rede Neural

```
classificador.fit(x_train, y_train, batch_size = 10, epochs = 1000)
```

### Predição com a Base de Dados para Teste

```
resultado = classificador.evaluate(x_test, y_test)
print(resultado)
```

```
previsoes = classificador.predict(x_test)
previsoes = previsoes > 0.5
print(previsoes)
```

```
y_test2 = [np.argmax(t) for t in y_test]
previsoes2 = [np.argmax(t) for t in previsoes]
```
