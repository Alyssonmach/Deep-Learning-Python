# Salvando o Classificador da Base de Dados Iris
#### Nesse arquivo será salvo o classificador da base de dados iris para posteriores usos.

### Importando as Bibliotecas

```python
# importando as bibliotecas do python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
from keras.models import model_from_json
```

### Carregando e Pré-processando os Dados

```python
# importando os dados no 'dataframe'
dataframe = pd.read_csv('iris.csv', encoding = 'utf-8', 
                        sep = ',')
# separando os atributos previsores e meta da base de dados
previsores = dataframe.iloc[:, 0:4].values
classe = dataframe.iloc[:, 4].values

# criando o objeto 'labelencoder'
labelencoder = LabelEncoder()

# realizando o escalonamento dos atributos meta
classe = labelencoder.fit_transform(classe)

# transformando as variáveis numéricas em variáveis do tipo dummy
classe_dummy = np_utils.to_categorical(classe)
```

![data-iris](Imagens/data-iris.png)

### Configurando a Rede Neural

```python
# criando o classificador para a rede neural multi-camada 
classificador = Sequential()

# adicionando a primeira camada densa 
classificador.add(Dense(units = 8, activation = 'relu',
                        kernel_initializer = 'normal', input_dim = 4))
# adicionando uma primeira camada de Dropout
classificador.add(Dropout(0.1))
# adicionando a segunda camada densa 
classificador.add(Dense(units = 8, activation = 'relu',
                        kernel_initializer = 'normal'))
# adicionando uma segunda camada de Dropout
classificador.add(Dropout(0.1))
# adicionando uma camada de saída
classificador.add(Dense(units = 3, activation = 'softmax'))
```

```python
# configurando alguns parâmetros da rede neural
classificador.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                      metrics = ['accuracy'])
```

![rede-neural](Imagens/rede-neural.PNG)

### Treinando a Rede Neural

```python
classificador.fit(previsores, classe_dummy, batch_size = 10, epochs = 2000)
```

```python
# observando a acurácia e a perda da rede neural feita
classificador.evaluate(previsores, classe_dummy)
```

### Salvando o Classificador

```python
# salvando o classificador em um arquivo .json
classificador_json = classificador.to_json()

# salvando os parâmetros da rede neural
with open('classificador_iris.json', 'w') as json_file:
  # escrevendo os dados no arquivo
  json_file.write(classificador_json)

# salvando os pesos configurados da rede neural
classificador.save_weights('classificador_iris.h5')
```
